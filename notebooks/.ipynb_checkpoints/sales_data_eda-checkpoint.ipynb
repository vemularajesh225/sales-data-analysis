```python
# sales_data_eda.ipynb

# 1. Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Load data
df = pd.read_csv('../data/raw_sales_data.csv')
df.head()
```

```python
# 3. Data Cleaning
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

```python
# Convert Order_Date to datetime
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
```

```python
# Create new columns
df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month
df['Quarter'] = df['Order_Date'].dt.quarter
df['Revenue'] = df['Units_Sold'] * df['Unit_Price']
```

```python
# Save cleaned data
df.to_csv('../data/cleaned_sales_data.csv', index=False)
```

```python
# 4. Top 10 Products by Revenue
top_products = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_products.values, y=top_products.index, palette="Blues_r", hue=top_products.index, legend=False)
plt.title('Top 10 Products by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Product')
plt.tight_layout()
plt.show()
```

```python
# 5. Top 5 Regions by Total Sales
top_regions = df.groupby('Region')['Revenue'].sum().sort_values(ascending=False).head(5)
top_regions.plot(kind='bar', title='Top 5 Regions by Revenue', figsize=(8,5), color='orange')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()
```

```python
# 6. Monthly Sales Trend Line
monthly_sales = df.groupby(['Year', 'Month'])['Revenue'].sum().reset_index()
monthly_sales['Period'] = pd.to_datetime(monthly_sales['Year'].astype(str) + '-' + monthly_sales['Month'].astype(str))
plt.figure(figsize=(12,6))
sns.lineplot(data=monthly_sales, x='Period', y='Revenue', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()
```

```python
# 7. Customer Retention Analysis (Optional)
retention = df.groupby('Customer_ID').size().value_counts().sort_index()
plt.figure(figsize=(8,5))
sns.barplot(x=retention.index, y=retention.values, palette="coolwarm", hue=retention.index, legend=False)
plt.title('Customer Retention - Purchase Frequency')
plt.xlabel('Number of Purchases')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.show()
```
